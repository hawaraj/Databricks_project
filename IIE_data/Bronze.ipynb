# Reading from CSV files and writing to bronze table


```python
df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/sales_details.csv"))
df.write.mode("overwrite").saveAsTable("workspace.bronze.sales_details")
   
df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/prd_info.csv"))
df.write.mode("overwrite").saveAsTable("workspace.bronze.prd_info")

df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/cust_info.csv"))
df.write.mode("overwrite").saveAsTable("workspace.bronze.cust_info")

df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/PX_CAT_G1V2.csv"))
df.write.mode("overwrite").saveAsTable("workspace.bronze.PX_CAT_G1V2")

df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/LOC_A101.csv"))
df=(spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/bronze/source_system/CUST_AZ12.csv"))
df.write.mode("overwrite").saveAsTable("workspace.bronze.LOC_A101")
```
